# Part A

```python
#Load the libraries
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Read the dataset
d = pd.read_csv('book.csv')
d
```

```python
d.mean()
```

```python
frequent_itemsets1 = apriori(d, min_support=0.05, use_colnames=True)
frequent_itemsets1
#a = pd.DataFrame(frequent_itemsets1.itemsets)
#b = len(a.itemsets)
#for i in a:
    #if b==(i):
        #continue
    #print(a)
```

```python
frequent_itemsets2 = apriori(d, min_support=0.1, use_colnames=True)
frequent_itemsets2
```

```python
frequent_itemsets3 = apriori(d, min_support=0.15, use_colnames=True)
frequent_itemsets3
```

```python
rules1 = association_rules(frequent_itemsets2, metric="confidence", min_threshold=0.5)
rules1
```

```python
rules2 = association_rules(frequent_itemsets2, metric="confidence", min_threshold=0.6)
rules2
```

```python
rules3 = association_rules(frequent_itemsets2, metric="confidence", min_threshold=0.65)
rules3
```

```python
rules1.sort_values('lift',ascending = False)
```

```python
rules2.sort_values('lift',ascending = False)
```

```python
rules3.sort_values('lift',ascending = False)
```

```python
rules1=pd.DataFrame(rules1[rules1.lift>1])
len(rules1)
```

```python
rules2=pd.DataFrame(rules2[rules2.lift>1])
len(rules2)
```

```python
rules3=pd.DataFrame(rules3[rules3.lift>1])
len(rules3)
```

```python
support = rules1['support'].values
confidence = rules1['confidence'].values
plt.title('Association Rules')
plt.xlabel('support')
plt.ylabel('confidence')
sns.regplot(x=support,y=confidence,fit_reg=False)
```

```python
support = rules2['support'].values
confidence = rules2['confidence'].values
plt.title('Association Rules')
plt.xlabel('support')
plt.ylabel('confidence')
sns.regplot(x=support,y=confidence,fit_reg=False)
```

```python
support = rules3['support'].values
confidence = rules3['confidence'].values
plt.title('Association Rules')
plt.xlabel('support')
plt.ylabel('confidence')
sns.regplot(x=support,y=confidence,fit_reg=False)
```

```python
support = rules1['support'].values
lift = rules1['lift'].values
plt.title('Association Rules')
plt.xlabel('support')
plt.ylabel('lift')
sns.regplot(x=support,y=lift,fit_reg=False)
```
